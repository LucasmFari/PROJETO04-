```python
import pandas as pd
import plotly.express as px


```

```python
junho2021 = pd.read_csv("https://raw.githubusercontent.com/LucasmFari/PROJETO04-/main/2021-06-gasolina-etanol%20(1).csv",on_bad_lines='skip', sep=';')

julho2021 = pd.read_csv("https://raw.githubusercontent.com/LucasmFari/PROJETO04-/main/2021-07-gasolina-etanol.csv",on_bad_lines='skip', sep=';')
```

```python
dados2021=junho2021.merge(julho2021, how='outer')
dados2021
```

```python

dados2021['Valor de Venda']= dados2021['Valor de Venda'].replace(',', '.', regex=True).astype(float)
dados2021
```

```python
dados2021['Data da Coleta'] = pd.to_datetime(dados2021['Data da Coleta'], format='%d/%m/%Y')
dados2021
```

##4) Qual o preço médio da gasolina e do etanol por estado?


```python
dados2021[['Produto', 'Estado - Sigla', 'Valor de Venda']].groupby(['Estado - Sigla', 'Produto']).mean().reset_index()
```

```python
rank = dados2021[['Produto', 'Estado - Sigla', 'Valor de Venda']].groupby(['Estado - Sigla', 'Produto']).mean().reset_index().sort_values(by='Valor de Venda')
fig = px.bar(rank, x='Estado - Sigla', y='Valor de Venda', color='Produto')
fig.show()
```
